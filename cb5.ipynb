<h1>Histogram</h1>

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
plt.xlabel("blood sugar levels")
plt.ylabel("No. of patients")
plt.title("Blood sugar relevence Data")
plt.hist(blood_sugar,bins=[80,100,125,150], rwidth=0.7,label="no=patients",color='red')
plt.legend(loc='best')

```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['#ADD8E6','pink'],label=['men','women'])
plt.legend()
```

```python
plt.ylabel("Sugar Level")
plt.xlabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['#ADD8E6','pink'],label=['men','women'],orientation='horizontal')
plt.legend()
```

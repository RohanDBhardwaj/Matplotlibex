<h1>Pie chart </h1>

```python
import matplotlib.pyplot as plt
```

```python
plt.title("Pie chart expences")
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
plt.show
```

```python
plt.pie(exp_vals,labels=exp_labels,autopct='%1.1f%%',radius=1.5,explode=[0,0.2,0,0,0.1])
plt.axis("equal")
plt.show()
plt.savefig('piecahart.pdf')
```

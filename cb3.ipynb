```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
days=[1,2,3,4,5,6,7]  
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]  
```

```python
min_value = min(min_t)
max_value = max(max_t)
plt.axhline(y=min_value, color='b', linestyle='-', label='Min Value')
plt.axhline(y=max_value, color='r', linestyle='--', label='Max Value')

plt.plot(days,max_t,alpha= 0.9)
plt.plot(days,min_t, alpha = 0.7)
plt.plot(days,avg_t,alpha= 1)
```

```python
#maximum an minimum
min_value = min(min_t)
max_value = max(max_t)
plt.axhline(y=min_value, color='b', linestyle='-', label='Min Value')
plt.axhline(y=max_value, color='r', linestyle='--', label='Max Value')
```

```python
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

plt.legend(loc='best')
```
